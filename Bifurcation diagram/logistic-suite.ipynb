# Logistic suite and bifurcation diagram

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from typing import List, Tuple

def logistic_map(u: float, r: float) -> float:
    return r * u * (1 - u)

def get_data(r: float, simple_size = 100) -> Tuple[List, List]:
    x = range(0, simple_size)
    u =  0.01
    y = []
    for i in x:
        u = logistic_map(u, r)
        y.append(u)
    return y

simple_size = 1000
x = []
y = []
for r in np.arange(-2, 4, 0.0005):
    y += get_data(r, simple_size)
    x += [r] * simple_size
```

```python
size = 16
fig = plt.figure(figsize = (20, 10))
plt.plot(x, y, '.', color="xkcd:blue",  markersize=0.05)
plt.text(-1, 1.50, r"$u_{n+1} = r \cdot u_n \cdot (1 - u_n), u_0 \in [0; 1], r \in [0, 4]$", fontsize=size)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.xlabel("r",  fontsize=size)
plt.ylabel("u",  fontsize=size)
# plt.savefig("bifurcation_diagram.png", bbox_inches='tight')
```
